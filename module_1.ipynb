# 1.8


```python
a = int(input())
list_inp = [-1] * a
k = 0
while n := input():
    if k < a:
        n = n.split()
        list_inp[k] = max(map(int, n))
        k += 1
list_inp.sort(reverse=True)
print(*list_inp, sep=';')
```

## Данный выше код не проходит проверку.
Ловим ошибку Run-time Error. Видимо по причине - заключительная строка не является пустой строкой.
Если ожидать ввода только ограниченного числа строк (не больше числа а), то не ловим ошибку.


```python
n = int(input())
list_inp = [-1] * n
for i in range(n):
    n_list = input().split()
    list_inp[i] = max(map(int, n_list))
list_inp.sort(reverse=True)
print(*list_inp, sep=';')
```

## Использование  stdin, также решает проблему 

```python
from sys import stdin


n = int(input())
lines = stdin.readlines()

list_inp = [-1] * n
for i in range(n):
    n_list = lines[i].split()
    list_inp[i] = max(map(int, n_list))
list_inp.sort(reverse=True)
print(*list_inp, sep=';')
```

```python
list_str = input().lower().split()
dict_ = {}
for word in list_str:
    dict_[word] = dict_.get(word, 0) + 1

key_m, max_ = ' ', 0
for key, value in dict_.items():
    if value > max_:
        max_ = value
        key_m = key
        
print(key_m, max_, sep=' ')
    
```

```python
list_str = input().lower().split()
set_str = set(list_str)
dict_ = {}
for word in set_str:
    count = list_str.count(word)
    dict_[count] = word

count_list = list(dict_.keys())
count_list.sort(reverse=True)
max_ = count_list[0]
        
print(dict_.get(max_, 0), max_, sep=' ')
```

```python
list_str = input().lower().split()
character = '!,.?;:#$%^&*()'
list_str = [''.join([char for char in word if (char not in character)]) for word in list_str]
list_str = [word for word in list_str if ((len(word) >= 5) & (len(set(word)) >= 4))]
dict_ = {}
for word in list_str:
    dict_[word] = dict_.get(word, 0) + 1
answer = [key for key, value in dict_.items() if value > 2]
answer.sort()
print(*answer, sep='\n')
```
